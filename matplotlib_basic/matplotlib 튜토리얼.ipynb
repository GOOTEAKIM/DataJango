### Pyplot 튜토리얼

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
# 하나만 좌표 쓰면 y 좌표

plt.ylabel("some numbers")
# y축 이름

plt.show()
```

```python
plt.plot([1,2,3,4,],
         [1,4,9,16])
# (1,1), (2,4), (3,9), (4,16) 의 점들을  직선으로 이음
```

```python
plt.plot([1,2,3,4,],
         [1,4,9,16], "ro")
# (1,1), (2,4), (3,9), (4,16) 의 좌표 값

plt.axis((0,6,0,20))
# x 좌표 : 0 ~ 6
# y 좌표 : 0 ~ 20

plt.show()
# 점만 찍음
```

```python
import numpy as np

t = np.arange(0.,5.,0.2)
# 0부터 5까지 0.2 간격으로 생성된 값들

plt.plot(t, t, 'r--', t, t*2, 'bs', t, t**3, 'g^')
# 3개의 데이터 플로팅

# 1. y = t (빨간색 점선)
# 2. y = 2t (파란색 사각형)
# 3. y = t^3 (초록색 삼각형) 

plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
# x축
values = [1,10,100]
# y축

plt.figure(figsize=(9,3))
# 그래프 크기 : 9인치 x 3인치

plt.subplot(131)
# 1행, 3열의 첫번째 그래프

plt.bar(names, values)
# bar 형태의 그래프

plt.subplot(132)
# 1행, 3열의 두번째 그래프

plt.scatter(names, values)
# 점 형태의 그래프

plt.subplot(133)
# 1행, 3열의 세번째 그래프

plt.plot(names, values)
# 선 형태의 그래프

plt.suptitle('Categorical Plotting')
# 전체 그래프 내용

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
    # 감쇠하는 진동 함수

t1 = np.arange(0.0, 5.0, 0.1)
# 0.0에서 5.0까지 0.1 간격의 시간 배열 생성
t2 = np.arange(0.0, 5.0, 0.02)
# 0.0에서 5.0까지 0.02 간격의 시간 배열 생성

# plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
# t1 : 파란색 점 ('bo') 표시, 
# t2 : 검은색 실선 ('k') 표시 

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
# cos 함수 빨간색 점선으로 표시

plt.show()
```

```python
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)  # arrange -> arange로 수정
s = np.cos(2*np.pi*t)
# 코사인 함수

line, = plt.plot(t, s, lw=2)
# lw == 선 두께

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             # (2,1) 위치에 주석 넣기
             # xytext=(3, 1.3) : 주석 텍스트가 표시될 지점
             # 그 지점에서 (2,1) 로 화살표로 잇는다.
             arrowprops=dict(facecolor='black', shrink=0.05),
             # 화살표 스타일
             )

plt.ylim(-2, 2)
# y축 : -2 ~ 2
plt.show()
```
